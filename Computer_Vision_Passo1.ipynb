```python
# Importe as bibliotecas necessárias
import cv2
from google.colab.patches import cv2_imshow

# Carregue uma imagem da internet
!wget -O imagem.jpg https://www.google.com/images/branding/googlelogo/2x/googlelogo_color_272x92dp.png
img = cv2.imread('/content/PingoCaipira.PNG')

# Exiba a imagem
cv2_imshow(img)

# Converta a imagem para escala de cinza
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Exiba a imagem em escala de cinza
cv2_imshow(gray)

# Aplique o detector de bordas Canny
edges = cv2.Canny(gray, 100, 200)

# Exiba a imagem com as bordas detectadas
cv2_imshow(edges)
```
